# Pandas case study
## Titanic
we will check data from ship data set

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ship = sns.load_dataset("titanic")
ship
```

```python
ship.head()
```

```python
# saving data fram einto csv file
ship.to_csv("ship.csv")
```

```python
# basic statistics / summary
ship.describe()
```

```python
ship.head()
```

```python
# droping few column and make new dataset
new_ship=ship.drop(["sibsp","parch"], axis=1)
new_ship.head()
```

```python
ship.mean()
```

```python
ship.groupby(["survived","age"]).mean()
```

```python
ship.groupby(["sex","class"]).mean()
```

```python
ship.value_counts(["survived"])
```

```python
ship.groupby(["sex"]).mean()
```

```python
ship[ship['age']<16].groupby(["sex", "class"]).mean()
```
